```python
import numpy as np

# Define the objective function
def f(x):
    return np.sin(x) - np.cos(x)

# Define the gradient of the objective function
def grad_f(x):
    return np.cos(x) + np.sin(x)

# Define the gradient descent algorithm with a stopping criterion
def gradient_descent(x_init, step_size, max_iter):
    x = x_init
    for i in range(max_iter):
        # Compute the gradient of the objective function at x
        grad = grad_f(x)
        
        # Update x using the gradient and step size
        x -= step_size * grad
        
        # Check the stopping criterion
        if np.abs(grad) < 1e-6:
            print(f'Stopping criterion met after {i} iterations')
            break
            
    return x

# Set the initial point, step size, and maximum number of iterations
x_init = 3.0
step_size = 0.05
max_iter = 500

# Run the gradient descent algorithm with the stopping criterion
x_opt = gradient_descent(x_init, step_size, max_iter)

# Print the optimal solution and objective value
print(f'Optimal solution: x = {x_opt:.4f}, f(x) = {f(x_opt):.4f}')
```
